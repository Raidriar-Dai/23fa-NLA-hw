# ECS 230 hw04 - polynomial fit plot

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_data_and_polynomial(data_file, coefficients):
    '''
    Example usage:
       coefficients = [a_0, a_1, ..., a_n] # Coefficients of the polynomial
       plot_data_and_polynomial('data.dat', coefficients)
    '''
    # Reading data
    data = np.loadtxt(data_file, skiprows=1)
    x, y = data[:, 0], data[:, 1]

    # Generating polynomial points
    X = np.linspace(min(x), max(x), 500)
    Y = sum(coefficients[i] * X**i for i in range(len(coefficients)))

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='red', label='Original Data Points')
    plt.plot(X, Y, color='blue', label='Polynomial Fit')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Polynomial Fit to Data Points')
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_data_and_multiple_polynomials(data_file, coefficients_list):
    '''
    Example usage:
       coefficients_list = [[a_0, a_1, ..., a_n], [b_0, b_1, ..., b_m], ...]
       plot_data_and_multiple_polynomials('data.dat', coefficients_list)
    '''
    # Reading data
    data = np.loadtxt(data_file, skiprows=1)
    x, y = data[:, 0], data[:, 1]

    # Determine the layout of the subplots
    rows = 2
    cols = 4

    # Create a figure and a set of subplots
    fig, axs = plt.subplots(rows, cols, figsize=(20, 10))  # Adjusted figure size

    # Flatten the array of axes
    axs = axs.flatten()

    # Loop through each set of coefficients and plot
    for i, coefficients in enumerate(coefficients_list):
        # Generating polynomial points
        X = np.linspace(min(x), max(x), 500)
        Y = sum(coefficients[j] * X**j for j in range(len(coefficients)))

        # Plotting on the subplot
        ax = axs[i]
        ax.scatter(x, y, color='red', label='Original Data Points')
        ax.plot(X, Y, color='blue', label=f'Polynomial Degree = {i+3}')
        ax.set_xlabel('X-axis')
        ax.set_ylabel('Y-axis')
        ax.set_title(f'Polynomial Degree = {i+3}')
        ax.legend()

    # Hide the unused subplot
    fig.delaxes(axs[-1])

    plt.tight_layout()
    plt.show()
```
